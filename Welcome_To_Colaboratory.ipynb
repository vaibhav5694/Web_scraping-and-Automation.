```
import requests
from bs4 import BeautifulSoup

# Define the URL of the Wikipedia page
url = 'https://en.wikipedia.org/wiki/COVID-19_pandemic_deaths'

# Send an HTTP GET request to the URL
response = requests.get(url)


```

```
# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')

```

```
rows_ = []
for table in soup.find_all('div', id = '2020-monthly-cumulative-covid-19-deaths'):
  for rows in table.find_all('tr'):
    rows_.append(rows.text)
```

```
final = []
for row in rows_:
  final.append(list (filter ( bool , row.splitlines()) ))
```

```
import pandas as pd
df = pd.DataFrame(data = final)
df
```
